```python
def decode_websocket_message(msg):
    msg_type = msg.split("|")[0]
    msg_data = "|".join(msg.split("|")[1:])

    if msg_type == "d":
        decode_type = msg_data.split(":")[0]
        data = ":".join(msg_data.split(":")[1:])

        arr = "".join(
            list(
                map(
                    lambda x: chr(ord(x[1]) + x[0] % 5),
                    enumerate(data),
                )
            )
        ).split("|")

        return msgArrayToDict(decode_type, arr)
    else:
        raise NotImplementedError


decode_websocket_message(
    "d|S:SESyRCAzPx1/z6..0z51.6z64.4z6-.2z6-.1z6-.0z.30/z.45/z2.0{7.*4{7.*5{7.*7{22,|1.-x13.yx|{0-x601-x91,.x90,/x8-/2533z553/z6-.3z.,|{z6-.4z.01852096,.x3/314913+/|"
)
```

```python
from enum import Enum


class TypeMsg(Enum):
    ACKNOWLEDGE = "a"
    CONNECTION = "c"
    DATA = "d"
    PING = "0"
    PONG = "1"
    REQUEST_SNAPSHOT = "ss"
    RESET = "r"
    SUBSCRIBE = "s"
    UNSUBSCRIBE = "u"
    VERSION = "v"


def sendMessage(typeMsg: TypeMsg, **kwargs):
    if typeMsg in [TypeMsg.ACKNOWLEDGE, TypeMsg.PONG]:
        return typeMsg.value

    if typeMsg in [TypeMsg.SUBSCRIBE, TypeMsg.UNSUBSCRIBE]:
        name = kwargs["name"]
        codes = kwargs["codes"]

        assert name is not None
        assert codes is not None
        return f"{typeMsg.value}|{name}:{','.join(codes)}"

    if typeMsg == TypeMsg.REQUEST_SNAPSHOT:
        type = kwargs["type"]

        if type in ["S", "D"]:
            codes = kwargs["codes"]

            return f"{typeMsg.value}|{type}:code={','.join(codes)}"

        elif type == "MI":
            codes = kwargs["codes"]

            return f"{typeMsg.value}|{type}:floorCode={','.join(codes)}"

        elif type == "CF":
            codes = kwargs["codes"]

            return f"{typeMsg.value}|{type}:floor={','.join(codes)}"

        else:
            raise NotImplementedError
```

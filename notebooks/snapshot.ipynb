```python
def decode(input):
    split = "".join(list(map(lambda x: chr(ord(x[1]) + x[0] % 5), enumerate(input)))).split("|")

    data = split[1:]

    if split[2] == "S" and split[3] == "10":
        return {
            "ma": data[0],
            "tc": data[3],
            "tran": data[5],
            "san": data[4],
            "tongkl": data[26],
            "tonggt": data[25],
            "muagia3": data[8],
            "muagia2": data[7],
            "muagia1": data[6],
            "muakl3": data[11],
            "muakl2": data[10],
            "muakl1": data[9],
            "gia": data[27],
            "kl": data[28],
            "bangia3": data[14],
            "bangia2": data[13],
            "bangia1": data[12],
            "bankl3": data[15],
            "bankl2": data[16],
            "bankl1": data[17],
            "cao": data[23],
            "thap": data[24],
            "nnm": data[21],
            "nnb": data[22],
        }
    elif split[2] == "ST" and split[3] in ["02", "03"]:
        return {
            "ma": data[0],
            "tc": data[3],
            "tran": data[5],
            "san": data[4],
            "tongkl": data[54],
            "tonggt": data[53],
            "muagia3": data[8],
            "muagia2": data[7],
            "muagia1": data[6],
            "muakl3": data[18],
            "muakl2": data[17],
            "muakl1": data[16],
            "gia": data[55],
            "kl": data[56],
            "bangia3": data[28],
            "bangia2": data[27],
            "bangia1": data[26],
            "bankl3": data[38],
            "bankl2": data[37],
            "bankl1": data[36],
            "cao": data[51],
            "thap": data[52],
            "nnm": data[49],
            "nnb": data[50],
        }
    else:
        print(split)
        raise NotImplementedError
```

# Market

```python
def marketCodeToName(code):
    if code == "10":
        return "VNINDEX"

    if code == "11":
        return "VN30"

    if code == "12":
        return "HNX30"

    if code == "13":
        return "VNXAllShare"

    if code == "02":
        return "HNX-INDEX"

    if code == "03":
        return "UPCOM"

    if code == "VN100":
        return "VN100"

    if code == "VNDIAMOND":
        return "VNDIAMOND"

    raise RuntimeError
```

## Market snapshot


```python
def decode(input):
    split = "".join(list(map(lambda x: chr(ord(x[1]) + x[0] % 5), enumerate(input)))).split("|")

    return {
        "name": marketCodeToName(split[1]),
        "somatang": split[4],
        "somatc": split[5],
        "somagiam": split[6],
        "gia": split[7],
        "tc": split[8],
        "high": split[9],
        "low": split[10],
        "kl": split[11],
        "gt": split[12],
    }
```
